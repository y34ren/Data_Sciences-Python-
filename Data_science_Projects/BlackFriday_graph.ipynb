```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(group,column,plot):
    g = df.groupby(group)[column].sum().sort_values(ascending=False)
    #print(g)
    pd.DataFrame(g).plot.bar(figsize=(15,9), rot=0)
   # plot(legend=False)
```

```python
df = pd.read_csv("./Project2Data/BlackFriday.csv")
df.fillna(0,inplace = True)
```

```python
df['Occupation'] = df['Occupation'].map({1: 'Surgeon', 2: 'Dentist', 3:'Nurse',0:'IT Manager',5:'Lawyer',6:'Sales Manager',
                                         7:'Financial Advisor',8:'Data Scientist',9:'Software Developer',10:'Teacher',11:'Engineer',
                                        12:'Web Developer',13:'Plumber',14:'Hairdresser',15:'Patrol Officer',16:'HR Specialist',
                                         17:'Paramedic',18:'Social Worker',19:'Translator',20:'Architecture',4:'Student'})
```

```python
plot('Occupation','Purchase', 'bar')
plt.xlabel('Occupation', fontsize=16)
plt.ylabel('Purchase', fontsize=16)
plt.legend(fontsize='xx-large')
```

```python
df['Marital_Status'] = df['Marital_Status'].map({0: 'Not Married', 1: 'Married'})
plot('Marital_Status','Purchase', 'bar')
plt.xlabel('Marital Status', fontsize=16)
plt.ylabel('Purchase', fontsize=16)
plt.legend(fontsize='xx-large')
```

```python
plot('Age','Purchase', 'bar')
plt.xlabel('Age', fontsize=16)
plt.ylabel('Purchase', fontsize=16)
plt.legend(fontsize='xx-large')
```

```python
plot('Gender','Purchase', 'bar')
plt.xlabel('Gender', fontsize=16)
plt.ylabel('Purchase', fontsize=16)
plt.legend(fontsize='xx-large')
```

```python
df['City_Category'] = df['City_Category'].map({'A': 'Atlanta', 'B': 'Toronto','C':"New York"})
plot('City_Category','Purchase', 'bar')
plt.xlabel('City', fontsize=16)
plt.ylabel('Purchase', fontsize=16)
plt.legend(fontsize='xx-large')
```

```python
df['Product_Category_1'] = df['Product_Category_1'].map({1: 'LG TV', 2: '3Ds', 3:'UGG',4:'Sneakers',5:'Iphone',6:'Mac Pro',
                                         7:'Bed',8:'PS4',9:'XBox',10:'Columbia',11:'Tiffiny',
                                        12:'Gift cards',13:'Movies',14:'Coach',15:'Kohl’s',16:'Sony TV',
                                         17:'Toys',18:'Tablets'})
plot('Product_Category_1','Purchase', 'bar')
plt.xlabel('Product Category', fontsize=16)
plt.ylabel('Purchase', fontsize=16)
plt.legend(fontsize='xx-large')
```
