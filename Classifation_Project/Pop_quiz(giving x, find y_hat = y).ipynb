```python
def OLS(y,y_hat):
    return (y - y_hat).dot(y - y_hat)

def R2 (y, y_hat):
    return 1 - OLS(y, y_hat) / OLS(y,y.mean())
```

```python
x = np.array((23,78,45,1,9,44,90,18,-21,12,63,32))
y = np.array((-21,90,1,78,9,44,23,32,45,18,63,12))

x, y= zip(*sorted(zip(x, y)))

X = []
x2 = np.array(x)
for i in x:
    X.append(np.exp(-(x2-i)**2))

X = np.array(X)

X = np.vstack([np.array([1]*len(X)), X]).T

w = np.random.randn(13)

# J is of OLS
J=[]

eta = 1e-1
epochs = 1000

for i in range ( epochs):
    y_hat2 = X.dot(w)
    J.append(OLS(y,y_hat2))
    w -= eta*X.T.dot(y_hat2 - y)

plt.figure(figsize= (12,9))

plt.plot(J)
```
