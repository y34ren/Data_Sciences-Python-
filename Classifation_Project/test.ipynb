```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sigmoid(h):
    return 1 / (1 + np.exp(-h))

def cross_entropy(y,p):
    return -np.sum(y*np.log(p) + (1 - y)*np.log(1 - p))

def accuracy(y,p):
    return np.mean(y == np.round(p))
```

```python
D = 2
K = 2
N = K*1000

X0 = np.random.randn((N//K),D) + np.array([3,1])
X1 = np.random.randn((N//K),D) + np.array([-1,1])
X = np.vstack((X0,X1))
y = np.array([0]*(N//K)+[1]*(N//K))

plt.figure(figsize=(9,6))
plt.axis([-5.0, 6.0, -5.0, 6.0])
plt.scatter(X[:,0],X[:,1],c = y,alpha=0.5)
```

```python
# Data prep
X = np.hstack((np.ones((N,1)),X))
```

```python
w = np.random.randn(D+1)

J=[]

eta = 1e-3

epochs = 1000

for i in range ( epochs):

    PHI = sigmoid(X.dot(w))
    J.append(cross_entropy(y,PHI))
    w -= eta*X.T.dot(PHI - y)


plt.plot(J)
```

```python
print ("Accuracy: {}".format(accuracy(y,PHI)))
```

```python
w2 = np.random.randn(D+1)

J2=[]

eta = 1e-3

epochs = 1000

for l1 in range (100):
    for i in range ( epochs):

        PHI2 = sigmoid(X.dot(w2))
        J2.append(((cross_entropy(y,PHI2) + l1*np.sum(np.abs(w2))),l1))
        w2 -= eta*(X.T.dot(PHI2 - y) + l1*np.sign(w2))


plt.plot(J)
```

```python
print ("Accuracy: {}".format(accuracy(y,PHI2)))
```
