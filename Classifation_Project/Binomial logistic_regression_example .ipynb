```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def J(y,p):    
    return -np.sum(y*np.log(p) + (1 - y)* np.log(1-p))
```

```python
D = 2
K = 2
N = K*1000

X0 = np.random.randn((N//K),D) + np.array([2,1])
X1 = np.random.randn((N//K),D) + np.array([-1,1])
X = np.vstack((X0,X1))

y = np.array([0]*(N//K)+[1]*(N//K))

plt.figure(figsize=(8,6))
plt.axis([-5.0, 5.0, -5.0, 5.0])
plt.scatter(X[:,0],X[:,1],c = y,alpha=0.5)
```

```python
# Data prep
X = np.hstack((np.ones((N,1)),X))
```

```python
#functions

def sigmoid(h):
    return 1/(1+np.exp(-h))


def accuracy(y,p):
    return np.mean(y == np.round(p))
```

```python
#code up getting outputs from input
w = np.random.randn(D+1)
#D+1 elements in w

p = sigmoid(X.dot(w))
```

```python
# Report Accuracy random untrained model

print ("Accuracy: {}".format(accuracy(y,p)))
```

```python
J=[]

eta = 1e-3

epochs = 1000

for i in range ( epochs):

    p = sigmoid(X.dot(w))
    J.append(-y.dot(np.log(p)) - (1 - y).dot(np.log(1 - p)))
    w -= eta*X.T.dot(p - y)


plt.plot(J)
```

```python

print ("Accuracy: {}".format(accuracy(y,p)))
```

```python
y_values = np.arange(-2,5)
x_values = -(w[0] + np.dot(w[2], y_values)) / w[1]
```
