```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 100

x = np.random.randn(N)

y = 4.0021 + 3.3321*x + np.random.randn(N)*0.7

plt.figure(figsize= (12,9))
plt.scatter(x,y)

```

```python
def OLS(y,y_hat):
    return (y - y_hat).dot(y - y_hat)

def R2 (y, y_hat):
    return 1 - OLS(y, y_hat) / OLS(y,y.mean())
```

```python
X = np.vstack([np.array([1]*N), x]).T
```

```python
# nomal equation solution

w = np.linalg.solve(X.T.dot(X), X.T.dot(y))

y_hat = X.dot(w)

plt.figure(figsize= (12,9))
plt.scatter(x,y)
plt.plot(x,y_hat, color = "red")

print ("R-Squared {}".format(R2(y,y_hat)))
```

```python
# Gradient Decent Model

w = np.random.randn(2)

# J is of OLS
J=[]

eta = 1e-3
epochs = 100

for i in range ( epochs):
    y_hat = X.dot(w)
    J.append(OLS(y,y_hat))
    w -= eta*X.T.dot(y_hat - y)

plt.figure(figsize= (12,9))

plt.plot(J)




```
