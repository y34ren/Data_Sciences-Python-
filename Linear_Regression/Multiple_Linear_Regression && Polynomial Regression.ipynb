```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

# Multiple Linear Regression

```python
N = 200
D = 2

X = np.random.randn(N,D)
x0 = np.array([[1]*N]).T
X = np.column_stack((x0, X))

X[:10,:]
```

```python
w_true = np.array([4.2,7.1,3.6])
y = X.dot(w_true) + np.random.randn(N)
```

```python
w = np.linalg.solve(X.T.dot(X),X.T.dot(y))
```

```python
y_hat = X.dot(w)
```

```python
fig = plt.figure(figsize = (12,9))

ax = fig.add_subplot(111, projection = "3d")

ax.scatter(X[:,1],X[:,2],y)
```

```python
#fig = plt.figure(figsize = (12,9))
#ax = fig.add_subplot(111, projection = "3d")
#ax.scatter(X[:,1],X[:,2],y)

#ax.plot_line(X,X,y_hat, color = "red")
```

```python
R2 = 1 - (y - y_hat).dot(y-y_hat) / (y - y.mean()).dot(y-y.mean())

print("R-Squared: {}".format(R2))
```

# Polynomial Regression  (non-linear relations)

```python
x = np.linspace(0,10,100)

y = 4.9 + 1.2 *x + 3.11456*(x**2) + np.random.randn(100)*6

plt.figure(figsize=(12,9))
plt.scatter(x,y)
```

```python
PBI = np.vstack((np.array([1]*100),x,x**2)).T
PBI[:10,:]
```

```python
w = np.linalg.solve(PBI.T.dot(PBI), PBI.T.dot(y))
y_hat = PBI.dot(w)
```

```python
plt.figure(figsize=(15,9))
plt.scatter(x,y)
plt.plot(x,y_hat, color = "red")
```

```python
R2 = 1 - (y - y_hat).dot(y-y_hat) / (y - y.mean()).dot(y-y.mean())

print("R-Squared: {}".format(R2))
```
